# 機器學習⼯作流程

```python
#機器學習的通⽤⼯作流程⼤致分為三個部分：

#1.定義任務了解問題領域和客⼾要求的業務邏輯。收集數據集，了解數據代表什麼，
#然後選擇衡量任務成功的⽅式。

#2.開發模型準備你的數據以便它可以被機器學習模型處理，選擇⼀個模型評估協議
#和⼀個簡單的基線來擊敗，訓練第⼀個具有泛化能⼒並且可以過擬合的模型，然後
#正則化和調整模型。


#3.部署模型向利益相關者展⽰您的⼯作，將模型發送到Web 服務器、移動應⽤程序
#、網⾴或嵌⼊式設備，監控模型在野外的性能，並開始收集您需要的數據構建下⼀代。
```

# 定義任務

```python
#您的模型將如何使⽤，以及它將如何適應您客⼾的業務流程？什麼樣的數據可⽤或可以
#收集？什麼樣的機器學習任務可以映射到業務問題。

#1.框定問題 : 
    #1.你的輸⼊數據是什麼？你想預測什麼？只有當你有可⽤的訓練數據時，你才能學會
    #預測某事
    #2.⾯臨什麼類型的機器學習任務？是⼆分類嗎？多類分類？
    #3.是否需要處理特定的限制條件？
#2.收集數據集 :
    #⼀旦您了解了任務的性質並知道您的輸⼊和⽬標將是什麼，就該進⾏數據收集了這是
    #⼤多數機器學習項⽬中最艱鉅、最耗時和最昂貴
#3.了解您的數據 :
    #將數據集視為⿊匣⼦是⾮常糟糕的做法。在開始訓練模型之前，您應該探索和可視化
    #您的數據，以深⼊了解是什麼使其具有預測性，這將為特徵⼯程和篩選潛在問題提供
    #信息。
#4.選擇衡量成功的標準
```

# 開發模型

```python
#1.準備數據 : 矢量化(浮點數)、值標準化(0-1)、處理缺失值
#2.選擇評估 : 
    #1.維護⼀個保留驗證集
    #2.進⾏ K 折交叉驗證
    #3.進⾏迭代 K折驗證
#3.擊敗基線 : 選擇正確的損失函數
#4 放⼤：開發⼀個過擬合的模型 : 
    #1. 添加圖層
    #2.使層更⼤
    #3. 訓練更多時期
#5.正則化和調整你的模型 : 
    #1.嘗試不同的架構；添加或刪除圖層
    #2.添加輟學
    #3.如果您的模型很⼩，請添加 L1 或 L2 正則化
    #4.嘗試不同的超參數（例如每層的單元數或優化器的學習率）以找到最佳配置
    #5.可選地，迭代數據管理或特徵⼯程：收集和註釋更多數據，開發更好的特徵
    #，或刪除似乎沒有信息的特徵
```

# 部署模型

```python
#1.向利益相關者解釋你的⼯作並設定期望
#2.發布推理模型
#3.在野外監控你的模型
#4.維護你的模型
```

# 總結

```python
#當你進⾏⼀個新的機器學習項⽬時，⾸先要定義⼿頭的問題 : 
    #1.了解更廣泛的背景 : 
        #你打算做什麼最終⽬標是什麼，限制是什麼?
    #2.收集和註釋數據集；確保您深⼊了解您的數據
    #3.選擇你將如何衡量你的問題的成功你將監控你的驗證數據的那些指標?
    
#⼀旦你理解了問題並且你有⼀個合適的數據集，開發⼀個模型：
    #1.準備好你的數據
    #2.是否需要驗證？
    #3.規範化您的模型並在驗證數據上調整其超參數性能
    
#當您的模型準備就緒並交付貨物時經測試的部署性能：
    #1.確保您設置適當對利益相關者的期望
    #2.優化推理的最終模型，並將模型發送到選擇的部署環境Web 服務器、移
    #動設備、瀏覽器、嵌⼊式設備等
    #3.監控模型在⽣產中的性能，並不斷收集數據，以便開發下⼀代模型
```
